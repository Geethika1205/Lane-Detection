```python
import cv2
import os

# Paths to both frame folders
input_frames_dir = "/content/adas_2000_frames"      # from adas input.mp4
output_frames_dir = "/content/adas11_2000_frames"   # from ADAS.mp4
combined_dir = "/content/combined_dataset"
os.makedirs(combined_dir, exist_ok=True)

# Total number of frames (assuming both folders have 2000)
num_frames = 2000

for i in range(num_frames):
    filename = f"frame_{i:05d}.jpg"
    input_path = os.path.join(input_frames_dir, filename)
    output_path = os.path.join(output_frames_dir, filename)

    if os.path.exists(input_path) and os.path.exists(output_path):
        input_img = cv2.imread(input_path)
        output_img = cv2.imread(output_path)

        # Resize both to same height if needed
        if input_img.shape != output_img.shape:
            height = min(input_img.shape[0], output_img.shape[0])
            input_img = cv2.resize(input_img, (input_img.shape[1], height))
            output_img = cv2.resize(output_img, (output_img.shape[1], height))

        # Combine side by side
        combined_img = cv2.hconcat([input_img, output_img])

        # Save combined image
        combined_path = os.path.join(combined_dir, filename)
        cv2.imwrite(combined_path, combined_img)

print(f"✅ Combined {num_frames} frames side-by-side into '{combined_dir}'")
```

```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Paths
source_dir = "/content/combined_dataset"
train_dir = "/content/data/train"
val_dir = "/content/data/val"

os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# Split 80-20
all_images = sorted(os.listdir(source_dir))
train_imgs, val_imgs = train_test_split(all_images, test_size=0.2, random_state=42)

# Move files
for img in train_imgs:
    shutil.copy(os.path.join(source_dir, img), os.path.join(train_dir, img))

for img in val_imgs:
    shutil.copy(os.path.join(source_dir, img), os.path.join(val_dir, img))

print(f"✅ Split into {len(train_imgs)} train and {len(val_imgs)} validation images.")
```

```python
from torch.utils.data import Dataset
from PIL import Image
import torchvision.transforms as transforms

class CombinedFrameDataset(Dataset):
    def __init__(self, folder, transform=None):
        self.folder = folder
        self.images = sorted(os.listdir(folder))
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_path = os.path.join(self.folder, self.images[idx])
        img = Image.open(img_path).convert('RGB')
        w, h = img.size
        input_img = img.crop((0, 0, w // 2, h))
        target_img = img.crop((w // 2, 0, w, h))

        if self.transform:
            input_img = self.transform(input_img)
            target_img = self.transform(target_img)

        return input_img, target_img
```

```python
from torch.utils.data import DataLoader

transform = transforms.Compose([
    transforms.Resize((256, 512)),
    transforms.ToTensor()
])

train_dataset = CombinedFrameDataset(train_dir, transform)
val_dataset = CombinedFrameDataset(val_dir, transform)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class SimpleCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1), nn.ReLU(),
            nn.MaxPool2d(2)
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(64, 64, 2, stride=2), nn.ReLU(),
            nn.Conv2d(64, 3, 3, padding=1)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
import torch.nn as nn
import torch.nn.functional as F

class UNetLite(nn.Module):
    def __init__(self):
        super().__init__()
        # Encoder
        self.enc1 = nn.Sequential(nn.Conv2d(3, 32, 3, padding=1), nn.ReLU())
        self.enc2 = nn.Sequential(nn.Conv2d(32, 64, 3, padding=1), nn.ReLU())
        self.pool = nn.MaxPool2d(2)

        # Bottleneck
        self.bottleneck = nn.Sequential(nn.Conv2d(64, 128, 3, padding=1), nn.ReLU())

        # Decoder
        self.up1 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.dec1 = nn.Sequential(nn.Conv2d(128, 64, 3, padding=1), nn.ReLU())

        self.up2 = nn.ConvTranspose2d(64, 32, 2, stride=2)
        self.dec2 = nn.Sequential(nn.Conv2d(64, 32, 3, padding=1), nn.ReLU())

        # Output
        self.out = nn.Conv2d(32, 3, 1)

    def forward(self, x):
        # Encode
        e1 = self.enc1(x)
        e2 = self.enc2(self.pool(e1))

        # Bottleneck
        b = self.bottleneck(self.pool(e2))

        # Decode
        d1 = self.up1(b)
        d1 = self.dec1(torch.cat([d1, e2], dim=1))

        d2 = self.up2(d1)
        d2 = self.dec2(torch.cat([d2, e1], dim=1))

        return self.out(d2)
```
