```python
import os
import zipfile
from PIL import Image
from tqdm import tqdm
import shutil

# Step 1: Unzip both archives
input_zip_path = "/content/Adas_Input_2000_frames.zip"
output_zip_path = "/content/ADAS_2000_frames.zip"
input_dir = "/content/input_frames"
output_dir = "/content/output_frames"
combined_dir = "/content/combined_dataset"

# Create directories
os.makedirs(input_dir, exist_ok=True)
os.makedirs(output_dir, exist_ok=True)
os.makedirs(combined_dir, exist_ok=True)

# Function to unzip files
def unzip(zip_path, extract_to):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_to)

# Unzip both zip files
unzip(input_zip_path, input_dir)
unzip(output_zip_path, output_dir)

# Step 2 & 3: Combine frames side-by-side
frame_names = sorted(os.listdir(input_dir))
for frame_name in tqdm(frame_names, desc="Combining frames"):
    input_path = os.path.join(input_dir, frame_name)
    output_path = os.path.join(output_dir, frame_name)

    if os.path.exists(input_path) and os.path.exists(output_path):
        input_img = Image.open(input_path)
        output_img = Image.open(output_path)

        # Ensure both images have the same height
        input_img = input_img.resize((input_img.width, input_img.height))
        output_img = output_img.resize((output_img.width, input_img.height))

        # Create new combined image
        combined_width = input_img.width + output_img.width
        combined_img = Image.new('RGB', (combined_width, input_img.height))
        combined_img.paste(input_img, (0, 0))
        combined_img.paste(output_img, (input_img.width, 0))

        # Save combined image
        combined_img.save(os.path.join(combined_dir, frame_name))

# Step 4: Zip the combined dataset
shutil.make_archive("/content/combined_dataset", 'zip', combined_dir)

print("✅ Combined dataset zip created at: /content/combined_dataset.zip")
```
