```python
import cv2
import os

# Input video path
video_path = "/content/adas input.mp4"

# Output directory to save frames
output_dir = "adas_2000_frames"
os.makedirs(output_dir, exist_ok=True)

# Open the video
cap = cv2.VideoCapture(video_path)

# Total number of frames in the video
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Calculate interval to get 2000 frames evenly
frame_interval = max(1, total_frames // 2000)

frame_count = 0
extracted = 0

while cap.isOpened() and extracted < 2000:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_count % frame_interval == 0:
        frame_filename = os.path.join(output_dir, f"frame_{frame_count:05d}.jpg")
        cv2.imwrite(frame_filename, frame)
        extracted += 1

    frame_count += 1

cap.release()
print(f"Extracted {extracted} frames to folder '{output_dir}'")
```

```python
import cv2
import os
import zipfile

# Optional: Unzip combined_dataset.zip if needed
zip_path = "combined_dataset.zip"  # Update if different name
unzip_dir = "combined_dataset"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(unzip_dir)
print(f"✅ Unzipped '{zip_path}' to '{unzip_dir}'")

# Frame extraction setup
video_path = "adas input.mp4"  # Make sure the video file is in the same directory
output_dir = "adas_2000_frames"
os.makedirs(output_dir, exist_ok=True)

# Open video
cap = cv2.VideoCapture(video_path)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Calculate how often to save a frame
frame_interval = max(1, total_frames // 2000)

frame_count = 0
extracted = 0

while cap.isOpened() and extracted < 2000:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_count % frame_interval == 0:
        frame_filename = os.path.join(output_dir, f"frame_{frame_count:05d}.jpg")
        cv2.imwrite(frame_filename, frame)
        extracted += 1

    frame_count += 1

cap.release()
print(f"✅ Extracted {extracted} frames to folder '{output_dir}'")
```

```python
import cv2
import os

# Paths to both frame folders
input_frames_dir = "/content/adas_2000_frames"      # from adas input.mp4
output_frames_dir = "/content/adas11_2000_frames"   # from ADAS.mp4
combined_dir = "/content/combined_dataset"
os.makedirs(combined_dir, exist_ok=True)

# Total number of frames (assuming both folders have 2000)
num_frames = 2000

for i in range(num_frames):
    filename = f"frame_{i:05d}.jpg"
    input_path = os.path.join(input_frames_dir, filename)
    output_path = os.path.join(output_frames_dir, filename)

    if os.path.exists(input_path) and os.path.exists(output_path):
        input_img = cv2.imread(input_path)
        output_img = cv2.imread(output_path)

        # Resize both to same height if needed
        if input_img.shape != output_img.shape:
            height = min(input_img.shape[0], output_img.shape[0])
            input_img = cv2.resize(input_img, (input_img.shape[1], height))
            output_img = cv2.resize(output_img, (output_img.shape[1], height))

        # Combine side by side
        combined_img = cv2.hconcat([input_img, output_img])

        # Save combined image
        combined_path = os.path.join(combined_dir, filename)
        cv2.imwrite(combined_path, combined_img)

print(f"✅ Combined {num_frames} frames side-by-side into '{combined_dir}'")
```
